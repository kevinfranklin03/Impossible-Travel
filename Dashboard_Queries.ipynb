```python
%sql
-- VISUALIZATION: Use "Counter" type
SELECT 
    COUNT(*) as total_alerts,
    ROUND(SUM(amount), 2) as total_fraud_amount,
    COUNT(DISTINCT card_id) as cards_impacted,
    MAX(speed_kmh) as max_speed_detected
FROM fraud_detection.raw.gold_fraud_alerts
WHERE alert_timestamp >= CURRENT_DATE()
```

```python
%sql
-- VISUALIZATION: Use "Line" type
SELECT 
    DATE_TRUNC('hour', alert_timestamp) as hour,
    severity,
    COUNT(*) as alert_count
FROM fraud_detection.raw.gold_fraud_alerts
WHERE alert_timestamp >= CURRENT_TIMESTAMP() - INTERVAL 24 HOUR
GROUP BY 1, 2
ORDER BY hour
```

```python
%sql
-- VISUALIZATION: Use "Map" type
SELECT 
    s.location,
    s.latitude,
    s.longitude,
    COUNT(*) as fraud_count,
    SUM(g.amount) as total_amount
FROM fraud_detection.raw.gold_fraud_alerts g
JOIN fraud_detection.raw.silver_all_transactions s
    ON g.transaction_id = s.transaction_id
GROUP BY 1, 2, 3
```
